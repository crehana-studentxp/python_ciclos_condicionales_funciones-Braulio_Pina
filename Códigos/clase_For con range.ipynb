```python
#ejemplo de range

#para numero en el rango de 0 al 11 -1 
for numero in range(0,11): #llega hasta el fin - 1 elemento
    
    #no utilizamos el elemento, solo tomamos la iteración que genera
    print(numero)
```

```python
#ejemplo de range

#para numero en el rango de 1 al 11 -1 
for numero in range(1,11): #puedo mover el punto inicial
    
    #no utilizamos el elemento, solo tomamos la iteración que genera
    print(numero)
```
